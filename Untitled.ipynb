```python
plt.figure(figsize=(10, 6))
plt.hist(df['Real Estate Value'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Real Estate Value')
plt.ylabel('Frequency')
plt.title('Distribution of Real Estate Values')
plt.show()
```

from the histogram we discover that the vast majority of data is in the first bar of the histogram suggesting that the value is highly skewed  with a long tail to the right. This is likly the result extreme outliers or a few very high-value transactions such as the purchace of premium and luxurios properties that cause the majority of the data to be clustered in lower value ranges.

to fix this we must further remove the outliers

```python
Q1 = df['Real Estate Value'].quantile(0.25)
Q3 = df['Real Estate Value'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers based on IQR
df_no_outliers = df[(df['Real Estate Value'] >= Q1 - 1.5 * IQR) & (df['Real Estate Value'] <= Q3 + 1.5 * IQR)]

# Plot histogram with adjusted data
plt.hist(df_no_outliers['Real Estate Value'], bins=20)
plt.show()
```

the new graph is much more normalized while still being naturally skewed

```python
num_rows = df_no_outliers.shape[0]
print("Number of rows:", num_rows)
```

using this method and simply removing these outliers resulted in a significant loss of data around 11% of data was removed and since these outliers are valid data points from real transaction and didnt resullt from measurements errors or bais i considered  not outright removing them instead replace them with the median and mean to standardize the data and avoid significant data loss
but since this data is heavily skewed we will use the median and not the mean

```python
Q1 = df['Real Estate Value'].quantile(0.25)
Q3 = df['Real Estate Value'].quantile(0.75)
IQR = Q3 - Q1


median = df['Real Estate Value'].median()

# Replace outliers with median
df_no_outliers = df.copy()  
df_no_outliers.loc[df['Real Estate Value'] < (Q1 - 1.5 * IQR), 'Real Estate Value'] = median
df_no_outliers.loc[df['Real Estate Value'] > (Q3 + 1.5 * IQR), 'Real Estate Value'] = median


plt.hist(df_no_outliers['Real Estate Value'], bins=20)
plt.show()
```

this replacement resulted in a heavy skew towards the bar around the median value which is an inncorrect intrupritation of the data and thus simply removing outliers is the better decision since normally its advised to have over 10 000 data points to conduct through study or research and our data stil excedes that even after the removal of outliers

```python
# finding out the Measures of Central Tendency will be done using mean and median
mean_value = df['Real Estate Value'].mean()
median_value = df['Real Estate Value'].median()

# Measures of Variability could be realised through the standard deviation and general range of the data
std_deviation = df['Real Estate Value'].std()
data_range = df['Real Estate Value'].max() - df['Real Estate Value'].min()

print("Measures of Central Tendency:")
print("Mean:", mean_value)
print("Median:", median_value)
print("\nMeasures of Variability:")
print("Standard Deviation:", std_deviation)
print("Range:", data_range)
```
