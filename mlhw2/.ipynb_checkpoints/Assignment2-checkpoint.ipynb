2) [Total 13 pts] 
- [1 pt] Load the `dataset.csv` (available on the same folder as this notebook) into a Pandas frame. Display the first 5 rows of the dataset.
- [12 pts] Preprocess the dataset by encoding the categorical variables using the `LabelEncoder` for the target variable and the `OneHotEncoder` for the features.
- Display the first 5 rows of the dataset after the preprocessing.

```python
import pandas as pd


df = pd.read_csv('dataset.csv')

print(df.head())
```

```python
#Preprocess the dataset by encoding the categorical variables using the LabelEncoder for the target variable and the OneHotEncoder for the features.
#first i will seperate the features from the target variable
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder



features = df.drop(columns=['Buy_iPhone'])
target = df['Buy_iPhone']

# label encoding target
label_encoder = LabelEncoder()
target_encoded = label_encoder.fit_transform(target)

# we are label encoding yes as 2 and no as 1 so im adding 1 to the result otherwise yes will be 1 and no will be 0 which looks similar to one hot encoding
target_encoded = target_encoded + 1 

# making a df for ptinting
target_encoded_df = pd.DataFrame({'Buy_iPhone_encoded': target_encoded})

# Display the first five rows of data set after preprocessing
print(target_encoded_df.head())

# OneHot Encoding features 
onehot_encoder = OneHotEncoder(drop='first')
#we are dropping first column of each feature which means we are doing k-1 encoding but we are not losing any data because 
#we can infer the values of the dropped column of the other columns(when they are all false means the dropped is true)
X_encoded = onehot_encoder.fit_transform(features)


X_encoded_dense = X_encoded.toarray()

encoded_feature_names = onehot_encoder.get_feature_names_out(features.columns)
X_encoded_df = pd.DataFrame(X_encoded_dense, columns=encoded_feature_names)

# Display the first five rows of data set after preprocessing
print(X_encoded_df.head())
```

2) [Total 13 pts]
- [3 pts] Split the dataset into 80% training set and 20% testing sets. 
- [3 pts] Train the model using the training set. 
- [3 pts] Plot the decision tree.
- [4 pts] Report the training accuracy and the testing accuracy of the trained decision tree.



```python
# 
# Spliting the data set
trainon, teston = train_test_split(df, test_size=0.2, random_state=42)

# Printing them
print("Shape of training set:", trainon.shape)
print("Shape of testing set:", teston.shape)
```

we get 50 rows for train and 13 rows for testing
50/63=79.3% which is almost 80% thus the splitting was correct

```python
#training
from sklearn.tree import DecisionTreeClassifier
X_train = trainon.drop(columns=['Buy_iPhone'])
y_train = trainon['Buy_iPhone']


# label encoding target
label_encoder = LabelEncoder()
y_trainencoded = label_encoder.fit_transform(y_train)

# making a df for ptinting
y_trainencoded_df = pd.DataFrame({'Buy_iPhone_Ytrain_encoded': y_trainencoded})

# Display the first five rows of data set after preprocessing
print(y_trainencoded_df.head())

# OneHot Encoding features 
onehot_encoder = OneHotEncoder(drop='first')
#we are dropping first column of each feature which means we are doing k-1 encoding but we are not losing any data because 
#we can infer the values of the dropped column of the other columns(when they are all false means the dropped is true)
X_trainencoded = onehot_encoder.fit_transform(X_train)


X_trainencoded_dense = X_trainencoded.toarray()

X_trainencoded_feature_names = onehot_encoder.get_feature_names_out(X_train.columns)
Xtrain_encoded_df = pd.DataFrame(X_trainencoded_dense, columns=X_trainencoded_feature_names)

print(Xtrain_encoded_df.head())





clf = DecisionTreeClassifier(random_state=42)

# Train the  model 
clf.fit(Xtrain_encoded_df, y_trainencoded_df)
```

```python
#plotting
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt



plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=Xtrain_encoded_df.columns, class_names=label_encoder.classes_)
plt.show()
```

```python
#evaluating
from sklearn.metrics import accuracy_score


train_predictions = clf.predict(Xtrain_encoded_df)


train_accuracy = accuracy_score(y_trainencoded_df, train_predictions)

print("Training Accuracy:", train_accuracy)

X_test = teston.drop(columns=['Buy_iPhone'])
y_test = teston['Buy_iPhone']


y_test_encoded = label_encoder.transform(y_test)


X_test_encoded = onehot_encoder.transform(X_test)


test_predictions = clf.predict(X_test_encoded)


test_accuracy = accuracy_score(y_test_encoded, test_predictions)

print("Testing Accuracy:", test_accuracy)
```

high training set accuracy which means the model doesnt suffer from underfitting and high testing set accuracy which mean the model
doesnt suffer from overfitting either, good overall results

3) [Total 9 pts] Evaluate the model using the testing set:

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

test_predictions = clf.predict(X_test_encoded)


conf_matrix = confusion_matrix(y_test_encoded, test_predictions)

class_labels = ["Yes", "No"]
#rearranging so the true positive is top left and true negative bottom right like the examples in the slides for better understanding
rearranged_conf_matrix = np.array([[conf_matrix[1, 1], conf_matrix[1, 0]], [conf_matrix[0, 1], conf_matrix[0, 0]]])

plt.figure(figsize=(8, 6))
sns.heatmap(rearranged_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

print(rearranged_conf_matrix)
#  classification report
class_report = classification_report(y_test_encoded, test_predictions)
print("Classification Report:")
print(class_report)
```

- [4 pts] Train a random forest model using the training set and evaluate it using the testing set. Fine-tune the number of trees used and the max_samples (try different numbers) and select the value that yield the best accuracy. Only include the best parameters in the final submittion.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# picked different numbers to try 
num_trees_values = [10, 50, 100] 
max_samples_values = [0.5, 0.75, 1] 

best_accuracy = 0
best_params = {}


for num_trees in num_trees_values:
    for max_samples in max_samples_values:
       
        rf_model = RandomForestClassifier(n_estimators=num_trees, max_samples=max_samples, random_state=42)
    
        rf_model.fit(Xtrain_encoded_df, y_trainencoded_df) 
        test_predictions = rf_model.predict(X_test_encoded)
        accuracy = accuracy_score(y_test_encoded, test_predictions)
  
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'num_trees': num_trees, 'max_samples': max_samples}

final_rf_model = RandomForestClassifier(n_estimators=best_params['num_trees'], 
                                         max_samples=best_params['max_samples'], 
                                         random_state=42)
final_rf_model.fit(Xtrain_encoded_df, y_trainencoded_df)  


final_test_predictions = final_rf_model.predict(X_test_encoded)

final_report = classification_report(y_test_encoded, final_test_predictions)
print("Best Parameters:", best_params)
print("Classification Report for Final Model:")
print(final_report)
```

```python
from sklearn.model_selection import cross_val_score


rf_model_cv = RandomForestClassifier(n_estimators=best_params['num_trees'], 
                                      max_samples=best_params['max_samples'], 
                                      random_state=42)


cv_scores = cross_val_score(rf_model_cv, Xtrain_encoded_df, y_trainencoded_df, cv=5)

# Printing
print("Cross-Validation Accuracies:", cv_scores)
print("Mean Accuracy:", cv_scores.mean())
print("Standard Deviation of Accuracy:", cv_scores.std())
```
